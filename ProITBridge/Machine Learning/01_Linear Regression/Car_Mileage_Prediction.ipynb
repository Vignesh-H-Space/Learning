## 4.Data Preparation

```python
cars_data
```

### There is nothing to prepare since because all features are numeric.

### Input Segregation

```python
X = cars_data[["HP","VOL","SP","WT"]]
```

```python
X
```

```python
y = cars_data["MPG"]
```

```python
y
```

## 5.Model Building

```python
linear_model = LinearRegression()
```

## 6.Model Training

```python
linear_model.fit(X,y)
```

```python
# Intercept - Where the line meets at the y - axis 
```

```python
linear_model.intercept_
```

```python
# Slope
```

```python
linear_model.coef_
```

## 7.Model Testing

```python
y_predicted = linear_model.predict(X)
y_predicted
```

## 8.Model Evaluation

```python
Error = y - y_predicted
Error
```

```python
cars_data["error"] = Error
```

```python
cars_data
```

## Homoscedacity Test

```python
plt.scatter(data=cars_data, x = "HP", y = "error")
```
