## Real world Analogy 1

* Using same function for Team members and Managers. 

* Team Members - Log in, View Tasks, Submit Tasks
* Managers - Log in, View Tasks, Assign Tasks


```python
# Define a function 

def Log_in(user_data):
    print(f"The {user_data['Name']} has logged in") 
    
def View_tasks(user_data):
    print(f"The {user_data['Name']} has viewed the tasks") 

def Submit_tasks(user_data):
    if user_data['Role'] != "Manager" :
            print(f"The {user_data['Name']} have submitted the tasks") 
    else :
            print(f"The {user_data['Name']} is a Manager and is not required to submit the tasks") 

def Assign_tasks(user_data):
    if user_data['Role'] == "Manager" :
            print(f"The {user_data['Name']} has assigned the tasks") 
    else :
            print(f"The {user_data['Name']} is a Team Member and cannot assign the tasks") 
    
# Data Collection 
user_data = {"Name":"Navin","Role":"Team Member"}

# Functions to be used 
Log_in(user_data)
View_tasks(user_data)
Submit_tasks(user_data)
Assign_tasks(user_data)
```

### With OOPS

```python
# Class creation
class Employee : 
    def __init__(self,name):
        self.name = name 
    
    def Log_in(self):
        print(f"The {self.name} has logged in") 
    
    def View_tasks(self):
        print(f"The {self.name} has viewed the tasks") 

# Class Inheritance 
class TeamMember (Employee):
    def Submit_tasks(self):
        print(f"The {self.name} have submitted the tasks") 

# Class Inheritance 
class Manager (Employee): 
    def Assign_tasks(self):
        print(f"The {self.name} has assigned the tasks")
```

```python
# Object Creation
# Here, john is the object 
```

```python
john = TeamMember("John")
nasir = Manager("Nasir")
```

```python
john.Log_in()
john.Submit_tasks()
john.View_tasks()
```

```python
nasir.Log_in()
nasir.Assign_tasks()
nasir.View_tasks()
```
