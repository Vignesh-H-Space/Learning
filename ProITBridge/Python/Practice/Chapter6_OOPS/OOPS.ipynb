* Using same function for Team members and Managers. 

* Team Members - Log in, View Tasks, Submit Tasks
* Managers - Log in, View Tasks, Assign Tasks


```python
# Define a function 

def Log_in(user_data):
    print(f"The {user_data['Name']} has logged in") 
    
def View_tasks(user_data):
    print(f"The {user_data['Name']} has viewed the tasks") 

def Submit_tasks(user_data):
    if user_data['Role'] != "Manager" :
            print(f"The {user_data['Name']} have submitted the tasks") 
    else :
            print(f"The {user_data['Name']} is a Manager and is not required to submit the tasks") 

def Assign_tasks(user_data):
    if user_data['Role'] == "Manager" :
            print(f"The {user_data['Name']} has assigned the tasks") 
    else :
            print(f"The {user_data['Name']} is a Team Member and cannot assign the tasks") 
    
# Data Collection 
user_data = {"Name":"Navin","Role":"Team Member"}

# Functions to be used 
Log_in(user_data)
View_tasks(user_data)
Submit_tasks(user_data)
Assign_tasks(user_data)
```

### With OOPS

```python
# Class creation
class Employee : 
    def __init__(self,name):
        self.name = name 
    
    def Log_in(self):
        print(f"The {self.name} has logged in") 
    
    def View_tasks(self):
        print(f"The {self.name} has viewed the tasks") 

# Class Inheritance 
class TeamMember (Employee):
    def Submit_tasks(self):
        print(f"The {self.name} have submitted the tasks") 

# Class Inheritance 
class Manager (Employee): 
    def Assign_tasks(self):
        print(f"The {self.name} has assigned the tasks")
```

```python
# Object Creation
# Here, john is the object 
```

```python
john.Log_in()
john.Submit_tasks()
john.View_tasks()
```

```python
nasir.Log_in()
nasir.Assign_tasks()
nasir.View_tasks()
```

## Abstraction 

```python
from abc import ABC, abstractmethod

# 1. This is now an Abstract Base Class (Blueprint)
class BankUser(ABC):
    
    @abstractmethod
    def deposit_cash(self):
        """Must be implemented by any specific bank user type"""
        pass

    @abstractmethod
    def cash_withdrawal(self):
        """Must be implemented by any specific bank user type"""
        pass

# 2. This is the Concrete Class (Actual Implementation)
class SecureUser(BankUser):
    def __init__(self, name, atm_pin):
        self.name = name
        self._atm_pin = atm_pin

    def get_pin(self, attempt):
        # Fixed: matches self._atm_pin from __init__
        return attempt == self._atm_pin 

    def deposit_cash(self):
        return f"{self.name} deposited cash"

    def cash_withdrawal(self):
        return f"{self.name} withdrew cash"

# Usage
user = SecureUser("Vivek", 5486)
print(user.deposit_cash())
```

## Inheritance

```python
# Parent Class (General)
class User:
    def __init__(self, name):
        self.name = name

    def deposit_cash(self):
        return f"{self.name} deposited cash"

# Child Class (Specialized) - Inherits from User
class SecureUser(User):
    def __init__(self, name, atm_pin):
        # super() calls the constructor of the Parent (User) class
        super().__init__(name)
        self._atm_pin = atm_pin  # Adds extra security attribute

    def get_pin(self, attempt):
        return attempt == self._atm_pin

    def cash_withdrawal(self):
        return f"{self.name} withdrew cash"

# Usage
useruh = SecureUser("Vivek", 5486)

# Accessing a method from the Parent class
print(useruh.deposit_cash())   # Vivek deposited cash

# Accessing a method from the Child class
print(useruh.cash_withdrawal()) # Vivek withdrew cash
```

## Polymorphism

```python
class User:
    def __init__(self, name):
        self.name = name

    def cash_withdrawal(self):
        return f"{self.name} withdrew cash"

class SecureUser(User):
    def __init__(self, name, atm_pin):
        # Pass ONLY name to the parent
        super().__init__(name) 
        # Handle atm_pin here locally
        self._atm_pin = atm_pin 

    def cash_withdrawal(self):
        return f"{self.name} (Secure ID: {self._atm_pin}) withdrew cash safely"

class PremiumUser(User):
    # This class doesn't have a custom __init__, 
    # so it automatically uses User's __init__
    def cash_withdrawal(self):
        return f"{self.name} withdrew cash with VIP perks!"

# This will now work perfectly:
users = [
    SecureUser("Vivek", 5486),
    PremiumUser("Ananya")
]

for user in users:
    print(user.cash_withdrawal())
```
