# Filtering, Selecting and Sorting 

```python
# Different functions of filtering, Selecting and Sorting
```

```python
import pandas as pd
```

```python
stats = pd.read_csv("Euro_2012_stats_TEAM.csv")
```

```python
print(stats)
```

```python
stats.columns  # Display all columns 
```

```python
# Filtering Data 
```

```python
filtered_Stats = stats[["Team","Goals","Red Cards","Yellow Cards"]]
```

```python
print(filtered_Stats)
```

```python
# Sort by most yellow cards
```

#### Top 5 Yellow Cards Team

```python
print(filtered_Stats.sort_values(by="Yellow Cards",ascending = False).head())
```

#### Top 5 Goals Team

```python
print(filtered_Stats.sort_values(by="Goals",ascending = False).head())
```

#### List teams who got Red Card 

```python
print(filtered_Stats[filtered_Stats["Red Cards"] > 0])
```

#### List teams who got more than 8 Yellow Cards

```python
print(filtered_Stats[filtered_Stats["Yellow Cards"] > 8])
```
