# Intermediate Data Visualisation

## Univariate Analysis

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
mt_cars = pd.read_csv("mtcars.csv")
```

```python
mt_cars
```

### Histogram 

```python
plt.hist(mt_cars.mpg)
plt.title("MPG Distribution")
plt.xlabel("MPG")
plt.ylabel("Frequency")
plt.show()
```

#### Difference between Histogram and Countplot 
#### Histogram - For Continous data
#### Countplot - For Discrete data

### Countplot 

```python
sns.countplot(mt_cars,x = "am")
plt.title("Automatic vs Manual Car Distribution")
plt.show()
```

```python
mt_cars["cyl"].value_counts()
```

### Pie Chart

```python
plt.pie(x = mt_cars["cyl"].value_counts(),labels = ["8 Cyl","4 Cyl","6 Cyl"],explode = [0.01,0.01,0.01],autopct = '%1.0f%%' )
plt.show()
```

## Bivariate Analysis

### Barplot

#### Discrete vs Continous - Bar Plot 
##### Discrete - x axis 
##### Continous - y axis 

```python
plt.bar(data = mt_cars,x = "am",height = "mpg")
```

```python
sns.barplot(data = mt_cars,x = "am",y = "mpg")
```

### Scatter Plot 

#### For two continuous data 

```python
plt.scatter(data = mt_cars, x = "wt", y = "mpg")
```

## Multivariate 

### Pair Plot

```python
sns.pairplot(data = mt_cars)
```
