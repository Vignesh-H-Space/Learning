# Data Analysis

## Reading TSV (Tab Separate Values) in python 

### Read CSV with delimiter

```python
import pandas as pd 
```

```python

# Importing a file and reading it using pandas 

pd.read_csv(filepath_or_buffer= "data.tsv.txt",sep = "\t")
```

### Read table function 

```python
order_details = pd.read_table(filepath_or_buffer= "data.tsv.txt")
```

```python
print(order_details)
```

```python
print(order_details.isnull()) # Entire table with true returned where value is null 

```

### Null function

```python
print(order_details.isnull().sum())  # Chain of Functions or Chain of Commands
# Proper Descripton of null is returned 
```

### Describe function

```python
order_details.describe()
```

```python
print(order_details.dtypes)  
```

```python
print(order_details.describe()) 
```

```python
print(order_details.describe(include="all"))
```

### Describe only particular Column

```python
print(order_details['order_id'].describe())
```
