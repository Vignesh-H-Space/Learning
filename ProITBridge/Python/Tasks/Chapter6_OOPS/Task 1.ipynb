## Abstraction , Inheritance & Polymorphism

### First, creating class with objects 

```python
class Player :
    def __init__(self,name,country):
        self.name = name
        self.country = country 
        
# Guess Player Name 
    def guess_player_name (self, guess):
        return guess.lower() == self.name.lower()

# Return the player country
    def get_country(self):
        return self.country

    
```

#### Object Creation 

```python
player1 = Player("Ronaldo","Portugal")
```

```python
player1.guess_player_name("Messi")
```

```python
player1.guess_player_name("Ronaldo")
```

```python
player1.get_country()
```

## Abstraction 

#### Abstract Class

```python
from abc import ABC, abstractmethod

class Player(ABC):

    @abstractmethod
    def guess_player_name (self, guess):
        pass
        
    @abstractmethod
    def get_country(self):
        pass
```

#### Inherited Class

```python
class CricketPlayer(Player):
    def __init__(self,name,country):
        self.country = country 
        
# Guess Player Name 
    def guess_player_name (self, guess):
        return guess.lower() == self.name.lower()

# Return the player country
    def get_country(self):
        return self.country
    
```

```python
user1 = CricketPlayer("Dhoni","India")
```

```python
user1.get_country()
```

## Inheritance 

#### Base Class 

```python
class Player :
    def __init__(self,name,country):
        self.name = name
        self.country = country 
        
# Guess Player Name 
    def guess_player_name (self, guess):
        return guess.lower() == self.name.lower()

# Return the player country
    def get_country(self):
        return self.country
```

#### Inherited Class

```python
class FootballPlayer (Player) :
    def get_country(self):
        return f"The player is from {self.country}"
    
```

```python
Person = FootballPlayer("Neymar","Brazil")
```

```python
Person.guess_player_name("nEyMaR")
```

```python
Person.get_country()
```

## Polymorphism

#### Base Class

```python
class Player :
    def __init__(self,name,country):
        self.name = name
        self.country = country 
        
# Guess Player Name 
    def guess_player_name (self, guess):
        return guess.lower() == self.name.lower()

# Return the player country
    def get_country(self):
        return self.country
```

#### Child Class 1 

```python
class FootballPlayer (Player) :
    def get_country(self):
        return f"The football player is from {self.country}"
    
```

#### Child Class 2 

```python
class CricketPlayer (Player) :
    def get_country(self):
        return f"The cricket player is from {self.country}"
```

```python
user1 = FootballPlayer("Pele","Brazil")
```

```python
user2 = CricketPlayer("Virat","India")
```

```python
user1.get_country()
```

```python
user2.get_country()
```

```python
user1.guess_player_name("Virat")
```

```python
user2.guess_player_name("Virat")
```
