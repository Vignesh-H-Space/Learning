### Merge function 

```python
import pandas as pd
```

```python
data1 = pd.DataFrame({'Key': ['A', 'B', 'C'], 'Value1': [1, 2, 3]})
data2 = pd.DataFrame({'Key': ['B', 'C', 'D'], 'Value2': [4, 5, 6]})
```

```python
merged_data = pd.merge(data1, data2, on='Key', how='inner')
print(merged_data)
```

```python
merged_data = pd.merge(data1, data2, on='Key', how='outer')
print(merged_data)
```

```python
merged_data = pd.merge(data1, data2, on='Key', how='left')
print(merged_data)
```

```python
merged_data = pd.merge(data1, data2, on='Key', how='right')
print(merged_data)
```

## Task 3 

### List Comprehension

### Syntax : new_list = [expression for item in iterable if condition]

#### Example 1 

```python
even_number = [x for x in range (20) if x % 2 == 0]
even_number
```

#### Example 2

```python
coordinates = [(x,y) for x in range (3) for y in range (3)]
```

#### Example 3

```python
data = ["even" if x % 2 == 0 else "odd" for x in range (20)]
data
```

#### Example 4

```python
data = [20,12,14,5,25]
category = ["high" if x > 20 else "medium" if x > 10 else "low" for x in data]
category
```

### Lambda Function

#### Map function

##### Example 1 

```python
list1 = ["apple","orange","watermelon"]
upper_list1 = list(map(lambda x: x.upper(),list1)) 
upper_list1
```

##### Example 2 

```python
list2 = [1, 6, 8, 3]
list3 = [5, 2, 9, 4]
add_list = list(map(lambda x,y: x + y , list2,list3))
add_list
```

### Filter Function

#### Example 1 

```python
list1 = ["apple","orange","watermelon"]
count_list1 = list(filter(lambda x :  len (x) < 7 , list1))
count_list1
```

#### Example 2

```python
list2 = [42,23,5,12,53,76,93]
even_list = list(filter(lambda x : x % 2 == 0 , list2))
even_list
```
