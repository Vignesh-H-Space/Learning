### Task - 2 : Even Numbers 

#### Create a list of numbers and print only even numbers using for loop 

```python
list1 = [int(x) for x in input("Enter numbers with space ").split()]

print (f"The list of numbers are : {list1}")
print("The even numbers present in the list are :",end = " ")
for val in list1:
    if val % 2 == 0 :
        print(val,end = " ")
```

### Task 3 - Division by Zero

#### A program which handles Division by zero using try-except

```python
print("Program to Divide two numbers")
num1 = int(input("Enter the first Number : "))
num2 = int(input("Enter the second Number : "))
try :
    print(num1/num2)
except ZeroDivisionError:       # Exception catch
    print("Error, Cannot divide by Zero")
```

### Task 4 - Vowels in a String

#### Given a String , count the number of vowels present 

```python
word = input("Enter a word: ")
counta = 0
print("The vowels are :",end=" ")
for i in word: 
    if i in 'aeiouAEIOU':
        counta = counta + 1
        print(i,end= " ")
print(f"\nThe total vowels are {counta}")
```

### Task 9 - Replace Missing Values in DataFrame

#### Replace all the missing values in Dataframe with column's mean

```python
import pandas as pd 

file = pd.read_csv("employee_data.csv")

# Before Change
print(file)

file = file.fillna(file.mean(numeric_only=True))        # Fill only numeric values since mean is only posible for numeric

# After Change
print(file)
```
