# Data Visualization

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
titanic_data = sns.load_dataset(name = "titanic")
titanic_data 
```

```python
plt.hist(titanic_data["class"])
plt.title("Passengers vs Class of Tickets")
plt.show()
```

```python
sns.countplot(data = titanic_data, x = "alive" )
plt.title("Total number of Survivers")
plt.show()
```

```python
# People who are alive 
```

```python
sns.barplot(data = titanic_data, x = "alive", y = "fare")
plt.title("Fare vs Survivors")
plt.show()
```

```python
sns.barplot(data = titanic_data, x = "alive", y = "adult_male")
plt.title("Survivors vs Adult Male")
plt.show()
```

```python
## This means the people who paid more fare where prioritized when it came to lifeboats 
```

```python
sns.histplot(data = titanic_data, x = "alive" , hue = "sex" , multiple = "stack")
plt.title("Sex vs Survivors")
plt.show()
```

```python
# So women were prioritised during rescue 
```

```python
sns.barplot(data=titanic_data, x="pclass", y="survived", hue="sex")
plt.title("Sex & Class vs Survivors")
plt.show()
```

```python
plt.pie( x = titanic_data["class"].value_counts(),labels = ["Third Class", "First Class", "Second Class"],explode = [0.01,0.01,0.01],autopct = '%1.0f%%' )
plt.title("Passenger Ratio")
plt.show()
```

```python
sns.countplot(data=titanic_data, x ="deck", hue="alive")
plt.title("Survived Passenger count in each deck")
plt.show()
```

```python
sns.stripplot(data=titanic_data, x="pclass", y="fare", hue="survived", dodge=True, alpha=0.5)
plt.title("Fare vs Survivors")
plt.show()
```
