## Probability Calculation for continuous random variable

### The length of human pregnancy is normally distributed with a mean of 272 days with a standard deviation of 9 days

* 1.State the Random variable

* 2.Find the probability of pregnancy lasting more than 280 days.

* 3.Find the probabilty of pregnancy lasting less than 250 days.

* 4.Find the probability that a pregnancy lasts between 265 and 280 days

```python
pip install scipy
```

```python
from scipy import stats 
```

#### Finding Probabilty of pregnancy lasting more than 280 days 

```python
1 - stats.norm.cdf(x = 280, loc = 272, scale = 9)
```

#### Finding Probabilty of pregnancy lasting less than 250 days 

```python
stats.norm.cdf(x = 250, loc = 272 , scale = 9)
```

#### Finding Probabilty of pregnancy between 265 and 280 days

```python
stats.norm.cdf(x = 280, loc = 272, scale = 9) - stats.norm.cdf(x = 265, loc = 272 , scale = 9)
```
