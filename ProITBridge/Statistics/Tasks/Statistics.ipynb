## 1. The historical average score in a certification exam is 65. A training institute introduces a new course. A random sample of 50 students trained under the new course has an average score of 68. At the 5% level of significance, test whether the new course improves performance.

```python
# H0 : Current process / Statement of No change - There is no significant difference in average score
# H1 : Statement of Change - The average score has increased 
```

```python
from scipy import stats
import numpy as np
```

```python
# Generating random array of numbers with mean as 68
```

```python
observations1 = np.random.normal(loc=68, scale=5, size=50).astype("int")
```

```python
observations1
```

```python
# ttest 1 sample with one sided 
```

```python
stats.ttest_1samp(a = observations1,popmean = 65, alternative = "greater")
```

```python
# From the tuple, taking only pvalue separately
```

```python
tstats,pvalue1 = stats.ttest_1samp(a = observations1,popmean = 65, alternative = "greater")
```

```python
pvalue1 
```

```python
# Coming to conclusion
```

```python
if pvalue1 < 0.05 :
    print("We reject H0")
else:
    print("We do not reject H0")
```

```python
# At 5% level of significance, we can reject the Null Hypothesis and claim that the average score has indeed increased
```

```python
# ttest 1 sample with one sided 
```

```python
 stats.ttest_1samp(a = observations2,popmean = 30,alternative = "greater")
```

```python
t_stats,pvalue2 = stats.ttest_1samp(a = observations2,popmean = 30,alternative = "greater")
```

```python
t_stats,pvalue2
```

```python
# Coming to conclusion
```

```python
if pvalue2 < 0.05 : 
    print("We Reject H0")
else:
    print("We do not reject H0")
```

```python
# At 5% level of significance, we can reject the Null Hypothesis and claim that the average resolution time has indeed increased
```
