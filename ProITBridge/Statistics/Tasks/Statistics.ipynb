## 2. A company claims that the average time taken to resolve customer tickets is 30 minutes. A random sample of 40 tickets shows a mean resolution time of 33 minutes. At the 5% level of significance, test whether the average resolution time has increased.

```python
# H0 : Current process / Statement of No change - There is no significant difference in average resolution time. 
# H1 : Statement of Change - There is a significant different in average resolution time. 
```

```python
from scipy import stats
import numpy as np
```

```python
observations = np.random.normal(loc=33, scale=5, size=40).astype("int")
```

```python
observations
```

```python
 stats.ttest_1samp(a = observations,popmean = 30,alternative = "greater")
```

```python
t_stats,pvalue = stats.ttest_1samp(a = observations,popmean = 30,alternative = "greater")
```

```python
t_stats,pvalue
```

```python
pvalue
```

```python
pvalue = pvalue / 2
```

```python
pvalue 
```

```python
if pvalue < 0.05 : 
    print("We Reject H0")
else:
    print("We do not reject H0")
```

```python
# At 5% level of significance, we can reject the Null Hypothesis and claim that the average resolution time has increased.
```
