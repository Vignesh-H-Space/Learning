## 1. The historical average score in a certification exam is 65. A training institute introduces a new course. A random sample of 50 students trained under the new course has an average score of 68. At the 5% level of significance, test whether the new course improves performance.

```python
# H0 : Current process / Statement of No change - There is no significant difference in average score
# H1 : Statement of Change - The average score has increased 
```

```python
from scipy import stats
import numpy as np
```

```python
# Generating random array of numbers with mean as 68
```

```python
observations1 = np.random.normal(loc=68, scale=5, size=50).astype("int")
```

```python
observations1
```

```python
# ttest 1 sample with one sided 
```

```python
stats.ttest_1samp(a = observations1,popmean = 65, alternative = "greater")
```

```python
# From the tuple, taking only pvalue separately
```

```python
tstats,pvalue1 = stats.ttest_1samp(a = observations1,popmean = 65, alternative = "greater")
```

```python
pvalue1 
```

```python
# Coming to conclusion
```

```python
if pvalue1 < 0.05 :
    print("We reject H0")
else:
    print("We do not reject H0")
```

```python
# At 5% level of significance, we can reject the Null Hypothesis and claim that the average score has indeed increased
```

## 3. A company claims that 60% of its users are satisfied with its product. In a survey of 200 users, 134 users report satisfaction. At the 5% level of significance, test whether customer satisfaction is higher than claimed.

```python
# H0 : Current process / Statement of No change - There is no significant difference in customer satisfaction
# H1 : Statement of Change - The customer satisfaction has increased
```

```python
from scipy import stats
import numpy as np
```

```python
# Since the mean is not given and its in percentage, we go with z-test
```

```python
# Given
p0 = 0.60
n = 200
x = 134

# Convert to "observations" (1 = satisfied, 0 = not satisfied)
observations3 = np.array([1]*134 + [0]*(200-134))

observations3
```

```python
# Sample proportion
p_hat = np.mean(observations3)

# Standard error
se = np.sqrt(p0 * (1 - p0) / n)

# Z-statistic
z_stat = (p_hat - p0) / se

# One-sided p-value
pvalue3 = 1 - stats.norm.cdf(z_stat)

z_stat, pvalue3
```

```python
if pvalue3 < 0.05:
    print("We Reject H0")
else:
    print("We do not reject H0")
```

```python
# At 5% level of significance, we can reject the Null Hypothesis and claim that the customer satisfaction has increased
```

## 4. Historically, 40% of applicants pass an entrance test. After a new preparation program, 90 out of 180 applicants pass. Test at the 5% level of significance whether the pass rate has improved.

```python
# H0 : Current process / Statement of No change - There is no significant difference in pass percentage
# H1 : Statement of Change - The average pass percentage has increased
```

```python
from scipy import stats
import numpy as np
```

```python
# Since the mean is in percentage, we use z-test
```

```python
# Given
p0 = 0.40
n = 180
x = 90

# Convert to "observations" (1 = satisfied, 0 = not satisfied)
observations4 = np.array([1]*90 + [0]*(180-90))

observations4
```

```python
# Sample proportion
p_hat = np.mean(observations4)

# Standard error
se = np.sqrt(p0 * (1 - p0) / n)

# Z-statistic
z_stat = (p_hat - p0) / se

# One-sided p-value
pvalue4 = 1 - stats.norm.cdf(z_stat)

z_stat, pvalue4
```

```python
if pvalue4 < 0.05:
    print("We Reject H0")
else:
    print("We do not reject H0")
```

```python
# At 5% level of significance, we can reject the Null Hypothesis and claim that the pass rate has increased.
```

## 5. Two teaching methods are used to train employees. A sample of 30 employees trained using Method A and 28 employees trained using Method B are tested. Test at the 5% level of significance whether there is a difference in average scores between the two methods.

```python
# Two tailed test
```

```python
# H0 : Null Hypothesis : Statement of no change, there is no significant change betweem the two methods
```

```python
# H1 : Alternate Hypothesis : Statement of change, there is a difference in averagew scores between two methods.
```

```python
from scipy import stats
import numpy as np
```

```python
np.random.seed(0)
```

```python
# Sample data 
```

```python
method_A = np.random.normal(loc=70, scale=8, size=30)
method_B = np.random.normal(loc=75, scale=8, size=28)
```

```python
method_A, method_B
```

```python
# Performing test
```

```python
t_test, pvalue = stats.ttest_ind(method_A, method_B)

t_test,pvalue
```

```python
if pvalue < 0.05:
    print("We Reject H0")
else:
    print("We do not reject H0")
```

```python
# At 5% level of significance, we do not reject the Null hypothesis and claim that there is no significant difference in average scores.
```
